```python
#@title Imagem em Escalada de Cinza
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
imgrgb = cv.imread('//content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 1/thepiratebay.jpg')
imgrgb = cv.cvtColor(imgrgb, cv.COLOR_BGR2RGB)
linha = imgrgb.shape[0]
coluna = imgrgb.shape[1]
canais = imgrgb.shape[2]
print("Imagem em Original")
plt.imshow(imgrgb);
plt.show()
imagem_cinza = np.zeros(( linha,coluna, canais), dtype='uint8' )
imagem_cinza = imgrgb.copy()
for x in range(linha):
  for y in range(coluna):
    imagem_cinza[x, y, :] = np.mean(imgrgb[x, y]) # R + G + B / 3
print("Imagem em Cinza");
plt.imshow(imagem_cinza);
plt.show()

```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
imgrgb = cv.imread('//content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 1/thepiratebay.jpg')
imgrgb = cv.cvtColor(imgrgb, cv.COLOR_BGR2RGB)
print("Imagem Original")
plt.imshow(imgrgb);
plt.show()
linha = imgrgb.shape[0]
coluna = imgrgb.shape[1]
canais = imgrgb.shape[2]
new_image = np.zeros(( linha,coluna, canais), dtype='uint8' )
constante = -1000
for x in range(linha):
  for y in range(coluna):
    r, g, b = imgrgb[x,y]
    if r + constante > 255:
       rsoma = 255
    if g + constante > 255:
       gsoma = 255
    if b + constante > 255:
       bsoma = 255
    if r + constante < 0:
       rsoma = 0
    if g + constante < 0:
       gsoma = 0
    if b + constante < 0:
       bsoma = 0
       new_image[x,y] = (rsoma,gsoma, bsoma)
plt.show()
print("Imagem Alterada")
plt.imshow(new_image);


```
