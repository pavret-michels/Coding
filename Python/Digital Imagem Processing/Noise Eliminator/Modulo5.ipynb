```python
#@title Eliminador de Ruído
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Carrega a imagem
img1A = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 5/Figura1A.png', 1)
# Aplica a técnica de Fechamento
kernel = np.ones((5, 5), np.uint8)
closing = cv.morphologyEx(img1A, cv.MORPH_CLOSE, kernel)

# Exibe a imagem original e o Resultado
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagem original
ax[0].imshow(cv2.cvtColor(img1A, cv2.COLOR_BGR2RGB))
ax[0].set_title("Imagem Original")

# Técnica de Fechamento
ax[1].imshow(closing, cmap='gray')
ax[1].set_title("Técnica de Fechamento")
plt.show()

# Carrega a imagem
img1B = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 5/Figura1B.png', 1)
# Aplica a técnica de Abertura
kernel = np.ones((5, 5), np.uint8)
abertura = cv.morphologyEx(img1B, cv.MORPH_OPEN, kernel)
# Exibe a imagem original e o Resultado
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagem original
ax[0].imshow(cv2.cvtColor(img1B, cv2.COLOR_BGR2RGB))
ax[0].set_title("Imagem Original")


# Técnica de Abertura
ax[1].imshow(abertura, cmap='gray')
ax[1].set_title("Técnica de Abertura")
plt.show()

# Carrega a imagem
img1C = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 5/Figura1C.png', 1)
#Aplicando a técnica de abertura e fechamento
kernel = np.ones((5, 5), np.uint8)
abertura = cv.morphologyEx(img1C, cv.MORPH_OPEN, kernel)
fechamentoEabertura = cv.morphologyEx(abertura, cv.MORPH_CLOSE, kernel)

# Exibe a imagem original e o Resultado
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagem original
ax[0].imshow(cv2.cvtColor(img1C, cv2.COLOR_BGR2RGB))
ax[0].set_title("Imagem Original")

# Técnica de Fechamento e Abertura
ax[1].imshow(fechamentoEabertura, cmap='gray')
ax[1].set_title("Técnica de Abertura e Fechamento")
plt.show()

```

```python
#@title Detector de Bordas
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem
img = cv2.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 3/Módulo 3B/imagem.png', 1)

# Converte a imagem para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplica o detector de bordas
edges = cv2.Canny(gray, 100, 200)

# Exibe a imagem original e o Resultado
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagem original
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Imagem Original")

# Bordas detectas
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Detecção de Bordas")

plt.show()
```

```python
#@title Transformada Top Hat
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem
img = cv2.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 5/arroz.png', 1)

# Aplica a transformada Top Hat
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (15, 15))
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

# Exibe a imagem original e a transformada Top Hat lado a lado
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Imagem original
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Imagem Original")

# Transformada Top Hat
ax[1].imshow(tophat, cmap='gray')
ax[1].set_title("Transformada Top Hat")

plt.show()
```
