```python
#@title Compacidade de um Quadrado
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Criação da imagem binária de um quadrado
size = 200
square = np.zeros((size, size), dtype=np.uint8)
cv2.rectangle(square, (50, 50), (150, 150), 255, -1)

# Função para calcular a compacidade
def compacidade(perimetro, area):
    return (perimetro ** 2) / area

# Mostrar imagem original
cv2_imshow(square)
print("Imagem Original")

# Calcular contornos, área e perímetro do quadrado original
contours, _ = cv2.findContours(square, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do quadrado original
compacidade_original = compacidade(perimetro, area)
print("Compacidade do quadrado original:", compacidade_original)

# Aplicar rotação
rotated_square = cv2.warpAffine(square, cv2.getRotationMatrix2D((size//2, size//2), 45, 1), (size, size))
cv2_imshow(rotated_square)
print("Imagem Rotacionada")

# Calcular contornos, área e perímetro do quadrado rotacionado
contours, _ = cv2.findContours(rotated_square, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do quadrado rotacionado
compacidade_rotacionada = compacidade(perimetro, area)
print("Compacidade do quadrado rotacionado:", compacidade_rotacionada)

# Aplicar translação
translation_matrix = np.float32([[1, 0, 50], [0, 1, 50]])
translated_square = cv2.warpAffine(square, translation_matrix, (size, size))
cv2_imshow(translated_square)
print("Imagem Transladada")

# Calcular contornos, área e perímetro do quadrado transladado
contours, _ = cv2.findContours(translated_square, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do quadrado transladado
compacidade_transladada = compacidade(perimetro, area)
print("Compacidade do quadrado transladado:", compacidade_transladada)

# Aplicar escala
scaled_square = cv2.resize(square, (size//2, size//2))
cv2_imshow(scaled_square)
print("Imagem Redimensionada")

# Calcular contornos, área e perímetro do quadrado redimensionado
contours, _ = cv2.findContours(scaled_square, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do quadrado redimensionado
compacidade_redimensionada = compacidade(perimetro, area)
print("Compacidade do quadrado redimensionado:", compacidade_redimensionada)
```

```python
#@title Triângulo Equilátero
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Função para calcular a compacidade
def compacidade(perimetro, area):
    return (perimetro ** 2) / area

# Criação da imagem binária de um triângulo equilátero
size = 200
triangle = np.zeros((size, size), dtype=np.uint8)
height = int(size * np.sqrt(3) / 2)
cv2.fillPoly(triangle, np.array([[(50, size-50), (size-50, size-50), (size//2, size-50-height)]], dtype=np.int32), 255)

# Exibe a imagem original
plt.imshow(triangle, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.show()

# Calcular contornos, área e perímetro do triângulo original
contours, _ = cv2.findContours(triangle, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do triângulo original
compacidade_original = compacidade(perimetro, area)
print("Compacidade do triângulo original:", compacidade_original)

# Aplicar rotação
rotation_matrix = cv2.getRotationMatrix2D((size//2, size//2), 45, 1)
rotated_triangle = cv2.warpAffine(triangle, rotation_matrix, (size, size))

# Calcular contornos, área e perímetro do triângulo rotacionado
contours, _ = cv2.findContours(rotated_triangle, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do triângulo rotacionado
compacidade_rotacionada = compacidade(perimetro, area)
print("Compacidade do triângulo rotacionado:", compacidade_rotacionada)

# Aplicar translação
translation_matrix = np.float32([[1, 0, 50], [0, 1, 50]])
translated_triangle = cv2.warpAffine(triangle, translation_matrix, (size, size))

# Calcular contornos, área e perímetro do triângulo transladado
contours, _ = cv2.findContours(translated_triangle, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do triângulo transladado
compacidade_transladada = compacidade(perimetro, area)
print("Compacidade do triângulo transladado:", compacidade_transladada)

# Aplicar escala
scaled_triangle = cv2.resize(triangle, (size//2, size//2))

# Calcular contornos, área e perímetro do triângulo redimensionado
contours, _ = cv2.findContours(scaled_triangle, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
area = cv2.contourArea(contours[0])
perimetro = cv2.arcLength(contours[0], True)

# Calcular e mostrar a compacidade do triângulo redimensionado
compacidade_redimensionada = compacidade(perimetro, area)
print("Compacidade do triângulo redimensionado:", compacidade_redimensionada)
```

```python
#@title Classificado de Imagens
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Função para calcular a compacidade
def compacidade(perimetro, area):
    return (perimetro ** 2) / area

# Tamanho da imagem
image_size = 300

# Cria uma imagem binária de fundo branco
triangle_image = np.ones((image_size, image_size), dtype=np.uint8) * 255
square_image = np.ones((image_size, image_size), dtype=np.uint8) * 255

# Desenha um triângulo equilátero na imagem
triangle_height = int(np.sqrt(3) / 2 * image_size)
triangle_base = image_size
triangle_x1 = 0
triangle_y1 = image_size
triangle_x2 = triangle_x1 + triangle_base
triangle_y2 = triangle_y1
triangle_x3 = triangle_x1 + triangle_base // 2
triangle_y3 = triangle_y1 - triangle_height
triangle_pts = np.array([[triangle_x1, triangle_y1], [triangle_x2, triangle_y2], [triangle_x3, triangle_y3]], np.int32)
cv2.fillPoly(triangle_image, [triangle_pts], 0)

# Desenha um quadrado na imagem
square_side = 100
square_x = (image_size - square_side) // 2
square_y = (image_size - square_side) // 2
square_image[square_y:square_y + square_side, square_x:square_x + square_side] = 0

# Calcula contornos e classifica os objetos
triangle_contours, _ = cv2.findContours(triangle_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
square_contours, _ = cv2.findContours(square_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for triangle_contour in triangle_contours:
    triangle_perimeter = cv2.arcLength(triangle_contour, True)
    triangle_area = cv2.contourArea(triangle_contour)
    triangle_compacidade = compacidade(triangle_perimeter, triangle_area)

    if triangle_compacidade <= (16 * np.pi / 3):
        cv2.drawContours(triangle_image, [triangle_contour], -1, (255, 0, 0), -1)  # Azul para triângulos

for square_contour in square_contours:
    square_perimeter = cv2.arcLength(square_contour, True)
    square_area = cv2.contourArea(square_contour)
    square_compacidade = compacidade(square_perimeter, square_area)

    if square_compacidade > (16 * np.pi / 3):
        cv2.drawContours(square_image, [square_contour], -1, (0, 255, 0), -1)  # Verde para quadrados

# Exibe as imagens classificadas e coloridas
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(triangle_image)
plt.title("Triângulo Equilátero Classificado")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(square_image)
plt.title("Quadrado Classificado")
plt.axis('off')

plt.show()
```
