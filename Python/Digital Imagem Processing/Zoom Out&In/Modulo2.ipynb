```python
#@title Zoom Out/In
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
imgrgb = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Aula 2/thepiratebay.jpg')
imgrgb = cv.cvtColor(imgrgb, cv.COLOR_BGR2RGB)
linha = imgrgb.shape[0]
coluna = imgrgb.shape[1]
print("Imagem Original")
plt.imshow(imgrgb);
plt.show()

print("Zoom In")
imgZIN = cv.resize(imgrgb,(2*linha, 2*coluna), interpolation = cv.INTER_CUBIC)
plt.imshow(imgZIN);
plt.show()

print("Zoom Out")
imgZout = cv.resize(imgrgb, (int(imgrgb.shape[1]*0.5), int(imgrgb.shape[0]*0.5)), interpolation=cv.INTER_CUBIC)
plt.imshow(imgZout);
plt.show()

```

```python
#@title Transforma√ß√£o s[k] = T(k, L, H, P, G) / Histograma
#import numpy as np
import cv2 as cv
%matplotlib inline
from matplotlib import pyplot as plt # Importa a blioteca pyplot
img = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Aula 2/Bordeaux.JPG',1)
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB); # Converte o formato da imagem de BGR para RGB
imgGray = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Aula 2/Bordeaux.JPG',0)
imgGray = cv.cvtColor(imgGray, cv.COLOR_BGR2RGB);
plt.imshow(imgRGB)
plt.xticks([]), plt.yticks([]) # esconde os valores das coordenadas X e Y
plt.title('Bordeaux - Original') # Nome da Imagem
plt.show()
#Imprimindo 2 imagens
print('Imprimindo 2 imagens')
fig, axs = plt.subplots(1,2)
axs[0].set_title('Bordeaux -RGB')
axs[0].imshow(imgRGB)
axs[1].set_title('Bordeaux - N√≠veis de Cinza')
axs[1].imshow(imgGray)
plt.show()
#salvar a imagem no drive (note as extens√µes dos arquivos - .jpg e .png)
cv.imwrite("/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Aula 2/Bordeaux1.JPG", imgGray)
#Calculo do histograma
histo = cv.calcHist([imgGray],[0],None,[256],[0,256])
plt.plot(histo)
plt.title('Histograma')
plt.show()
plt.hist(imgGray.ravel(),256,[0,256]); #ravel() - transforma a imagem em um vetor.
plt.title('Histograma')
plt.show()
print(histo[100])
print(histo.shape)
#Equaliza√ß√£o do Histograma
imgG = cv.cvtColor(img, cv.COLOR_RGB2GRAY);
imgGray = cv.cvtColor(imgGray, cv.COLOR_BGR2RGB);
eq = cv.equalizeHist(imgG);
plt.title('Imagem Equalizada')
plt.imshow(cv.cvtColor(eq, cv.COLOR_BGR2RGB))
plt.show()
plt.hist(eq.ravel(),256,[0,256]); #ravel() - transforma a imagem em um vetor.
plt.title('Histograma Equalizado')
plt.show()
print(histo[100])
print(histo.shape)
print("")
```
