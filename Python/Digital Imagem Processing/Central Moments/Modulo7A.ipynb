```python
import cv2
import numpy as np
import pandas as pd

from google.colab.patches import cv2_imshow

# Carregar uma imagem de exemplo (pode ser substituída por outra imagem)
image = cv2.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 6/modelo.jpg', cv2.IMREAD_GRAYSCALE)

# Calcular os momentos de Hu
moments = cv2.moments(image)
hu_moments = cv2.HuMoments(moments)

# Criar uma lista de nomes para os momentos
moment_names = ['Momento ' + str(i) for i in range(7)]

# Criar um DataFrame para armazenar os momentos
data = {'Momentos': moment_names, 'Valor': hu_moments.flatten()}
df = pd.DataFrame(data)

# Exibir a tabela
print("Sete Momentos Centrais Normalizados e Invariantes:")
print(df)

# Exibir a imagem original
cv2_imshow(image)



```
