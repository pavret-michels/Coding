```python
#@title Calcula área e perímetro do Objeto
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Relação de pixels para centímetros
relacao_px_cm = 2.54

# Carrega a imagem de referência e converte para escala de cinza
referencia = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 3/Módulo 3C/Referencia.jpeg', 0)

# Carrega a imagem do novo objeto
objeto = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 3/Módulo 3C/Objeto1.png', 1)

# Converte a imagem do novo objeto para escala de cinza
objeto_gray = cv.cvtColor(objeto, cv.COLOR_BGR2GRAY)

# Redimensiona a imagem do novo objeto para ter o mesmo tamanho da imagem de referência
objeto_gray_resized = cv.resize(objeto_gray, (referencia.shape[1], referencia.shape[0]))

# Calcula a diferença absoluta entre a imagem do novo objeto e a imagem de referência
diff = cv.absdiff(objeto_gray_resized, referencia)

# Cria uma imagem binária, onde os pixels diferentes são marcados como brancos
thresh = cv.threshold(diff, 0, 255, cv.THRESH_BINARY)[1]

# Encontra os contornos na imagem binária
contours, _ = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

# Calcula a área do objeto contando o número de pixels brancos na imagem binária
area_objeto_px = cv.countNonZero(thresh)
area_objeto_cm = area_objeto_px * relacao_px_cm**2

# Calcula o perímetro do objeto
perimetro_objeto = 0
for contour in contours:
    perimetro_objeto += cv.arcLength(contour, True)

# Exibe a área e o perímetro do objeto em centímetros
print("Área do objeto:", area_objeto_cm, "cm²")
print("Perímetro do objeto:", perimetro_objeto * relacao_px_cm, "cm")


```
