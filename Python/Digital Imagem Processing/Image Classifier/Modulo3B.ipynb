```python
#@title Classificador de Imagem
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Carrega a imagem as imagens
img = cv.imread('/content/drive/MyDrive/Mestrado/Processamento Digital de Imagens/Módulo 3/Módulo 3B/imagem.png', 1)
cv2_imshow(img)

#Converter a imagem para escala de cinza
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# aplicar a limiarização 
ret, thresh = cv.threshold(gray, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)

# Procura os contorno na imagem
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

#Indica o maior e menor contorno
largest_contour = max(contours, key=cv.contourArea)
smallest_contour = min(contours, key=cv.contourArea)

#Pinta as manchas
cv.drawContours(img, contours, -1, (255, 0, 0), 2)
cv.drawContours(img, [largest_contour], -1, (0, 0, 255), 3)
cv.drawContours(img, [smallest_contour], -1, (0, 255, 255), 3)

#Exibe a nova Imagem
cv2_imshow(img)
```
